```python
import pandas as pd
```

```python
world_cup = pd.read_csv(r'womens-world-cup.csv')
world_cup
```

```python
world_cup['continent'] = ''
world_cup
```

```python
world_cup['squad'].value_counts()
```

# HinzufÃ¼gen der Konitnente

```python
africa = world_cup.loc[(world_cup['squad'] == 'Nigeria') | (world_cup['squad'] == 'Cameroon') | 
                        (world_cup['squad'] == 'Ghana') |  
                        (world_cup['squad'] == 'CÃ´te d\'Ivoire') | 
                        (world_cup['squad'] == 'Equ. Guinea') | 
                        (world_cup['squad'] == 'Jamaica') | 
                        (world_cup['squad'] == 'South Africa') ]
africa
```

```python
africa = africa.assign(continent= 'Afrika')
africa
```

Europa

```python
asien = world_cup.loc[(world_cup['squad'] == 'Japan') | 
                        (world_cup['squad'] == 'China PR') |  
                        (world_cup['squad'] == 'Chinese Taipei') | 
                        (world_cup['squad'] == 'Korea DPR') | 
                        (world_cup['squad'] == 'Korea Rep') | 
                        (world_cup['squad'] == 'Russia') | 
                        (world_cup['squad'] == 'Thailand') ]
asien
```

```python
namerika = world_cup.loc[(world_cup['squad'] == 'USA') | 
                        (world_cup['squad'] == 'Canada') |  
                        (world_cup['squad'] == 'Mexico') | 
                        (world_cup['squad'] == 'Costa Rica') ]
namerika
```

```python
frames = [africa, asien, europe, samerika, namerika, oceanien]


result = pd.concat(frames)
result
```
